```python
# Handling divide by zero using try and except block

num=int(input("Enter the numerator: "))
den= int(input("Enter the denominator: "))
try:
    ans=num/den
    print(ans)
except:
    print("This is Divide by zero error: ")
    den= int(input("Enter the denominator: "))

ans=num/den
print(ans)
```
