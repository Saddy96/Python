```python
# Creating a new database test 
# SQLite3 can be integrated with Python using sqlite3 module, which was written by Gerhard Haring.
# https://www.sqlite.org/index.html
import sqlite3
# Connect to the test2 database
conn = sqlite3.connect('test2.db')
print("Opened database successfully")
print(conn.total_changes)
conn.close()
```

```python
# Creating a table in the test database
import sqlite3
conn = sqlite3.connect('test.db')
p='CREATE TABLE Student(ID INT PRIMARY KEY  NOT NULL,NAME  TEXT  NOT NULL, AGE  INT  NOT NULL, ADDRESS CHAR(50), PROGRAM REAL)'
conn.execute(p)
print("Table created successfully")
conn.commit()
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect('test.db')
c1=conn.execute("Select * from Student")
print(c1.fetchall())
conn.commit()
```

```python
# Insert into the data base
import sqlite3
conn = sqlite3.connect('test.db')
conn.execute("INSERT INTO Student (ID,NAME,AGE,ADDRESS,PROGRAM) \
      VALUES (16, 'Ananya', 23, 'Bangalore', 12 )");
conn.execute("INSERT INTO Student (ID,NAME,AGE,ADDRESS,PROGRAM) \
      VALUES (25, 'Allen', 25, 'Mysore', 12 )");
c1 = conn.execute("Select * from Student")
print(c1.fetchall())
conn.commit()
conn.close()
```

```python
# Selecting rows from the table
conn = sqlite3.connect('test.db')
c=conn.execute("Select * from Student")
for col in c:
   print("ID = ", col[0])
   print("NAME = ", col[1])
   print("AGE= ", col[2])
   print("ADDRESS = ", col[3]) 
print("Operation done successfully")
conn.close()
```

```python
# Conditional selection 
conn = sqlite3.connect('test.db')
c1=conn.execute("Select * from Student where NAME= 'Ananya'")
print(c1.fetchone())   # Fetches one record
```

```python
# Conditional selection 
conn = sqlite3.connect('test.db')
c1=conn.execute("Select * from Student where NAME= 'Ananya'")
print(c1.fetchmany(3))
```

```python
# Update Operation
import sqlite3
conn = sqlite3.connect('test.db')
print("Opened database successfully")
conn.execute("UPDATE Student set ADDRESS = 'ZX' where ID = 11")
conn.commit()
conn.close()
```

```python
# Deleting a record
import sqlite3
conn = sqlite3.connect('test.db')
conn.execute("DELETE from Student where Name= 'Ananya';")
conn.commit()
print("Total number of rows deleted :", conn.total_changes)
c=conn.execute("Select * from Student")
print(c.fetchall())
conn.close()
```

```python
# And OR Clause
import sqlite3
conn = sqlite3.connect('test.db')
c=conn.execute("SELECT * FROM Student WHERE AGE = 25 OR PROGRAM= 13")
print(c.fetchall())
conn.close()
```

```python
#  SQLite LIKE operator is used to match text values against a pattern using wildcards.
#The percent sign (%)
##The underscore (_)
import sqlite3
conn = sqlite3.connect('test.db')
c=conn.execute("SELECT * FROM Student WHERE AGE LIKE '2%'")
print(c.fetchall())
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect('test.db')
c=conn.execute("SELECT * FROM Student WHERE NAME LIKE '_l%'") 
print(c.fetchall())
conn.close()
```

```python
# Limit and Offset
# offset will leave the no of records mentioned in bloew case 1
import sqlite3
conn = sqlite3.connect('test.db')
c=conn.execute("SELECT * FROM Student LIMIT 3 OFFSET 1")
print(c.fetchall())
conn.close()
```

```python
# Order By Single Column
import sqlite3
conn = sqlite3.connect('test.db')
c=conn.execute("SELECT * FROM Student ORDER BY  AGE DESC")
print(c.fetchall())
conn.close()
```

```python
# Create table for referencing from another table
s="CREATE TABLE supplier_groups (group_id integer PRIMARY KEY,group_name text NOT NULL)"
conn = sqlite3.connect('test.db')
conn.execute(s)
conn.commit()
conn.close()
```

```python
p= "CREATE TABLE suppliers (supplier_id INTEGER PRIMARY KEY,supplier_name TEXT NOT NULL ,group_id INTEGER, FOREIGN KEY (group_id) REFERENCES supplier_groups (group_id) ON UPDATE SET NULL ON DELETE SET NULL)"
conn = sqlite3.connect('test.db')
conn.execute(p)
conn.commit()
conn.close()
```
