```python
numbers = [1, 2, 3, 4, 5]
sq = []
def squ(no):
    for num in no:
        sq.append(num * 2)
    print(sq)

   
squ(numbers)
```

```python


# Doing the same functionality using MAP
def squ(number):
     return number*2
numbers = [1, 2, 3, 4, 5]

#map(function, iterable) 
squ = map(squ, numbers)
print(list(squ))
```

```python
# Computing total of two list using map

PWP = [23,45,56,87,98]
SMR = [45,65,78,34,76]

def total(PWP,SMR):
    return(PWP+SMR)
list(map(total,PWP,SMR))
```

```python
def is_even(x):
    if x % 2 == 0:
        return True
    else:
        return False
f = filter(is_even, [1, 3, 10, 45, 6, 50])

for i in f:
    print(i)
```

```python
def is_odd(x):
    if x % 2 != 0:
        return True
    else:
        return False
f = filter(is_odd, [1, 3, 10, 45, 6, 50])

for i in f:
    print(i)
```

```python
def cons(x):
    if x in ['a','e','i','o','u']:
        return True
    else:
        return False
f = filter(cons, ['a','d','e','g'])

for i in f:
    print(i)
```

```python
from functools import reduce
def add(x,y):
    return x+y
def multiply(x,y):
    return x *y
data=[1,2,3,4]
print(reduce(add,data))
reduce(multiply,data)
```

```python
def fact(n):
    return reduce(lambda p,r: p*r, range(1,n+1),1)
fact(3)
```
