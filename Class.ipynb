```python
# POLYMORPHISM

class Square:
    
    def __init__(self,side):
        self.side=side
    
    def calculate_area_sq(self):
        return self.side * self.side
class Rectangle:
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def calculate_area_rect(self):
        return self.l * self.b
    
class Triangle:
    def __init__(self,b,h):
        self.b=b
        self.h=h
    def calculate_area_tri(self):
        return 0.5 * self.b * self.h
sq = Square(3)
rect = Rectangle(3,5)
print("Area of square: ", sq.calculate_area_sq())
print("Area of Rectangle: ", rect.calculate_area_rect())

tri = Triangle(5,10)
print("Area of triangle: ", tri.calculate_area_tri())
```

```python
# Operator overloading
# Write a Dunder method for the same.

import math
 
class Circle:
 
    def __init__(self, radius):
        self.radius = radius
 
    def get_result(self):
        return self.radius
 
    def area(self):
        return math.pi * self.radius ** 2
 
    def __add__(self, another_circle):
        return Circle(self.radius + another_circle.radius)
 
    def __sub__(self, another_circle):
        return Circle(self.radius - another_circle.radius)

# Different ways are there to to do operator overloading like __add__, __mul__ etc
    def __mul__(self, another_circle):
        return Circle(self.radius * another_circle.radius)
 
    
 
c1 = Circle(10)
print(c1.get_result())
print(c1.area())
 
c2 = Circle(15)
print(c2.get_result())
print(c1.area())
 
c3 = c1 + c2
print(c3.get_result())
 
c4 = c1 * c2
print(c4.get_result())

c5 = c2 - c1
print(c5.get_result())
```

```python
# Area of shape based on method overloading

class Shape:
      
    def calculate_area(self,inp1,inp2=None):
        if inp2 is not None:
            return inp1 * inp2
        else:
            return inp1 * inp1

sq = Shape()
print("Area of square: ", sq.calculate_area())

```

```python
# INHERITENCE
# Parent class and Child class

# Inheritance in Python
# If __init__ of base class in not called in the inherited class, it will not be invoked
# Sub-classing 
# Calling a constructor of the parent class by mentioning the parent class name in the declaration of the child class 
# is known as sub-classing.
#A child class identifies its parent class by sub-classing.
class Student:    # parent class, super class
  
              
        def __init__(self, name, idnumber):   
                self.name = name
                self.idnumber = idnumber
        def display(self):
                print(self.name)
                print(self.idnumber)
  
# child class , derived class
class Specialization(Student):           
        def __init__(self, name, idnumber, sname, year):
                self.sname = sname
                self.year = year
  
                # invoking the __init__ of the parent class 
                Student.__init__(self, name, idnumber) 
        
        def __str__(self):
            return '{},{},{},{}'.format(self.name,self.idnumber,self.sname,self.year)
        
        
       
# creation of an object variable or an instance
s = Specialization('Rahul', "95CP215", "BA", 1)  
s.display()
   
# calling a function of the class Person using its instance
print(s) 
```

```python
class Person:  
    #defining constructor  
    def __init__(self, personName, personAge):  
        self.name = personName  
        self.age = personAge  
  
    #defining class methods  
    def showName(self):  
        print(self.name)  
  
    def showAge(self):  
        print(self.age)  
  
    #end of class definition  
  
# defining another class  
class Student: # Person is the  
    def __init__(self, studentId):  
        self.studentId = studentId  
  
    def getId(self):  
        return self.studentId  
  
  
class Resident(Person, Student): # extends both Person and Student class  
    def __init__(self, name, age, id):  
        Person.__init__(self, name, age)  
        Student.__init__(self, id)  
        
    def __str__(self):
        return '{},{},{}'.format(self.name,self.age,self.studentId)
        
  
  
# Create an object of the subclass  
r1 = Resident('John', 30, '102')  
print(r1)
```

```python
# Mulitlevel Inheritance
# When a child class becomes a parent class for another child class.
class Family:
   
    def show_family(self):
        print("This is our family:")
 
 
# Father class inherited from Family
class Father(Family):
    def __init__(self,fathername):
        self.fathername=fathername
 
    
 
# Mother class inherited from Family
class Mother(Family):
     def __init__(self,mothername):
            self.mothername=mothername
 
     
 
# Son class inherited from Father and Mother classes
class Daughter(Father, Mother):
    def __init__(self,myname,fathername,mothername):
        self.myname=myname
        Father.__init__(self,fathername) 
        Mother.__init__(self,mothername) 
        
    
    def __str__(self):
            return '{},{},{}'.format(self.myname,self.fathername,self.mothername)
             
 
s1 = Daughter("Mamta","Raj","Tripti")  # Object of Son class
s1.show_family()
print(s1)
```
