```python
def fact(x):
    factVal = 1
    for i in range(x, 0, -1):
        factVal = i * factVal
    return factVal

no = int(input('Enter the no '))
factResult = fact(no)
print('factorial val ', factResult)

print('Squaring the result')
print(factResult**2)
```

```python
def sum(x):
    sumVal = 1
    for i in range(2, x+1):
        sumVal = sumVal + i
    return sumVal

def fact(y):
    factVal = 1
    for i in range(y, 0, -1):
        factVal = i * factVal
    return factVal

no = int(input('Enter the no '))
selection = int(input('Enter the selection '))

if selection == 1:
    sumResult = sum(no)
    print('sum Value ', sumResult)
elif selection == 2:
    factResult = fact(no)
    print('factorial val ', factResult)
else:
    print('Invalid Selection Entry')
```

```python
def details():
    name=input("Enter the name: ")
    dep=input("Enter Dept name: ")
    sp=input("Enter specialization: ")
    marks = int(input("Enter marks: "))
    return name, dep, sp, marks

def marksComparision(listData):
    highestMarks = max(listData)
    lowestMarks = min(listData)

    indexhighest = listData.index(highestMarks)
    indexlowest = listData.index(lowestMarks)
    return indexhighest, indexlowest

   
p=[]
q=[]
r=[]
s=[]

re=int(input("How many records you want to enter"))
for i in range(0,re): 
    p1,q1,r1,s1=details()
    p.append(p1)
    q.append(q1)  
    r.append(r1)
    s.append(s1)

averageMarks = (sum(s) / len(s))
print('Average Marks ', round(averageMarks,2))

index1,index2 = marksComparision(s)
print('user name who got highest marks is ',p[index1])
print('user name who got lowest marks is ',p[index2])
```

```python
# A recursive function is a function that calls itself

def factorial(n):
    """This is a recursive function
    to find the factorial of an integer"""
    if n == 1:
        return 1
    else:
        return (n * factorial(n-1))

num = int(input('Enter the no '))
print("The factorial of", num, "is", factorial(num))

# Call happens from top to down whereas result will be passed from bottom to top
```
