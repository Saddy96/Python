```python
# Different Ways of creating a Dictionary

# empty dictionary
my_dict = {}
my_dict = dict()

# dictionary with integer keys {key:value,}
my_dict = {1: 'apple', 2: 'ball', 3:'ball'}
print(my_dict)

# dictionary with mixed keys
my_dict = {'name': 'John', 1: [2, 4, 3]}
print(my_dict)

my_dict = dict([(1,'apple'), (2,'ball')])
print(my_dict)
```

```python
# Keys have to be unique in dictionary 
dictionary = {"key_1": "value_1", "key_1": "value_2", "key_2": "value_2"}
print(dictionary)
```

```python
# Adding items in the dictionary
my_dict = {1: 'apple', 2: 'ball', 3:'ball'}
print(my_dict)
my_dict[4]="pear"
print(my_dict)
print(len(my_dict))
```

```python
my_dict1 = {1: 'apple', 2: 'ball', 3:'ball'}
print(my_dict1)
my_dict1.update({4:"pear"})
print(my_dict1)
```

WAP to check if the key exist in the dictionary or not. If it does not exist, enter the key and its corresponding value in the dictionary and print it 

```python
my_dict = {1: 'apple', 2: 'ball', 3:'bat'}

key_check = int(input("Enter the value to be checked "))

if key_check in my_dict.keys():
    print("Already Exist")
            
else:
    print("No Value exist")
    val = input("Enter the value to be updated: ")
    my_dict[key_check] = val

print(my_dict)
```

#WAP to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x*x).

```python
my_dict = {}

size = int(input('Enter the size of dict : '))

count = 1
while count <= size:
    key = count
    value = count * count
    my_dict[key] = value
    count+=1
    
print(my_dict)  
```

#Adding new names to dictionary and if it already exist increase the counter




```python
# removing items from dictionary using pop() and popitem()
my_dict = {1: 'apple', 2: 'ball', 3:'bat'}
a = my_dict.pop(1)
del my_dict[2]
print(a)
print(my_dict)
#popitem() removes the last inserted item from the dictionary
my_dict = {1: 'apple', 2: 'ball', 3:'bat'}
b = my_dict.popitem()
print(b)
print(my_dict)
```

```python
# making a copy of dictionary using '='
my_dict = {1: 'apple', 2: 'ball', 3:'bat'}
my_dict1=my_dict
print(my_dict1)
my_dict[4]='money'
print(my_dict1)
print(my_dict)

# making a copy using copy()
my_dict = {1: 'apple', 2: 'ball', 3:'bat'}
my_dict1=my_dict.copy()
my_dict1.clear()
print(my_dict1)
print(my_dict1)
print(my_dict)
```

```python
# Creating a dictionary from two lists
# Using  zip() function
# Zip creates an iterator  that will aggregate elements from two or more iterables.
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
sampleDict = dict(zip(keys, values))
print(sampleDict)
```
