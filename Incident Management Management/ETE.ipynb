```python
# Module required for DB Connection
import sqlite3
from tkinter import *
```

```python
# Create a DB

# Below code will create a DB if not there
def tableCreation():
    conn = sqlite3.connect('incidentMS.db')
    print("Opened database successfully")

    query='''CREATE TABLE login(
                        userID TEXT PRIMARY KEY  NOT NULL,
                        PASSWORD TEXT  NOT NULL
                        )'''
    query1='''CREATE TABLE ticktes(
                        raisedBy TEXT NOT NULL,
                        asignedTo TEXT  NOT NULL,
                        ticketIssue TEXT  NOT NULL,
                        ticketStatus TEXT  NOT NULL
                        
                        )'''
    conn.execute(query)
    conn.execute(query1)
    
    print("All required Tables are created successfully")
    conn.commit()

    conn.close()

tableCreation()
```

```python
def getUser(userName, Password):
    conn = sqlite3.connect('incidentMS.db')
    query="SELECT count(*) FROM login WHERE userID = '{username}' AND PASSWORD = '{Password}';".format(username = userName, Password = Password)
    c1 = conn.execute(query)
    successVal = 0
    for col in c1:
        if (col[0] == 1):
            print("SUCCESSFULLY VERIFIED")
            successVal = 1
        else:
            print("INVALID CREDENTIALS")

    conn.commit()
    conn.close()

    return successVal

def createuser(userName, Password):
    conn = sqlite3.connect('incidentMS.db')
    query="INSERT INTO login (userID, PASSWORD) VALUES ('"+userName+"', '"+Password+"');"

    conn.execute(query)
    print("Added the Entry")
    conn.commit()
    conn.close()

def updateTicket(raisedBy,asignedTo):
    conn = sqlite3.connect('incidentMS.db')
    query="UPDATE ticktes SET ticketStatus = 'Completed' WHERE raisedBy = '{}' AND asignedTo ='{}';".format(raisedBy,asignedTo)

    c2 = conn.execute(query)

    conn.commit()
    conn.close()
    
updateTicket()
#createuser('Ram', 'Admin') 
#getUser('Ram','Admin')
#insertTickets('Kim','Ram','Require New Software','Completed')
#getTickets('Ram')
```

```python
class employee:
    def __init__(self, userName): 
        self.userName = userName
    
    def showUserName(self):
        print(self.userName)

# child class , derived class        
class hubEmployee(employee):
    
    def __init__(self, userName):
        
        # invoking the __init__ of the parent class
        employee.__init__(self, userName)
        self.user = userName
        self.raisedBy = []
        self.asignedTo = userName
        self.ticketIssue = []
        self.ticketStatus = []
        
    def process(self):
        print("\nWhat do you want to do")
        count = 1
        listData = ["GET TICKETS","ADD TICKETS", "QUIT THE PROCESS"]
        for i in listData:
            print("Choose ",count, "for", i)
            count += 1
    
    # Based on the process function the below function will ask the user to enter their chosen option
    def processSelection(self):
        while(True):
            try:
                selVal = int(input("Enter the selection from above options"))
                if selVal == 1 or selVal == 2 or selVal == 3 or selVal == 4:
                    print("You have selected ",selVal)
                    return selVal
                    break;
                else:
                    print("Please selct only from the above option")
            except ValueError:
                print("\n Not a valid entry, try again.")
            
    def getTickets(self, userName):
        conn = sqlite3.connect('incidentMS.db')
        query="SELECT * FROM ticktes WHERE asignedTo = '{username}';".format(username = userName)
        c2 = conn.execute(query)
        print(c2.fetchall())

        conn.commit()
        conn.close()

    def insertTickets(self,raisedBy,asignedTo,ticketIssue,ticketStatus):
        conn = sqlite3.connect('incidentMS.db')
        query="INSERT INTO ticktes (raisedBy,asignedTo,ticketIssue,ticketStatus) \
          VALUES ('{}', '{}', '{}','{}');".format(raisedBy,asignedTo,ticketIssue,ticketStatus)

        c2 = conn.execute(query)

        conn.commit()
        conn.close()
    
    def updateTicket(self,raisedBy,asignedTo):
        conn = sqlite3.connect('incidentMS.db')
        query="UPDATE ticktes SET ticketStatus = 'Completed' WHERE raisedBy = '{}' AND asignedTo ='{}';".format(raisedBy,asignedTo)

        c2 = conn.execute(query)

        conn.commit()
        conn.close()
       
```

```python
# Main Functions
def main():
    while(True):
        print("*"*45)
        print("WELCOME TO INCIDENT MANAGEMENT SYSTEM")
        print("*"*45)
        st = ""
        while(True):
            sel = input("Are you registred user, Enter y for YES or n for NO: ")
            if sel.lower() == 'y' or sel.lower() == 'n':
                st = "Continue"
                break;
            else:
                print("Invalid selection entry")
        
        if st == "Continue":
            if sel.lower() == 'y':
                userId = input("Enter your Username: ")
                password = input("Enter your Password: ")
                successVal = getUser(userId,password)
                
                if successVal == 1:
                    user = hubEmployee(userId)
                    user.process()
                    engName = user.userName
                    processVal = user.processSelection()
                    if processVal == 1:
                        user.getTickets(engName)
                    
                    elif processVal == 2:
                        print("\nAdding New Tickets  ")
                        raisedBy = input("Enter the name of the person ticket raised for: ")
                        asignedTo = input("Enter the person name to assign this ticket to: ")
                        ticketIssue = input("Enter the ticket issue: ")
                        ticketStatus = input("Enter the ticket status: ")
                        user.insertTickets(raisedBy,asignedTo,ticketIssue,ticketStatus)
                    
            else:
                userId = input("Enter your Username: ")
                password = input("Enter your Password: ")
                addition = createuser(userId, password)
                
main()                 
```

```python
def mainGUI():
    global main_screen
    main_screen = Tk()
    main_screen.geometry("700x550")
    main_screen.title("INCIDENT MANAGEMENT SYSTEM")
    Label(main_screen, text="LOGIN SCREEN", bg="Blue", width="300", height="2", font=("Calibri", 13)).pack(fill="x")
    Label(main_screen, text="").pack()
    
    # create a Form label 
    Label(text="Choose Login Or Register", bg="Red", width="300", height="2", font=("Calibri", 13)).pack() 
    Label(text="").pack() 

    # create Login Button 
    Button(text="Login", height="2", width="30").pack() 
    Label(text="").pack() 

    # create a register button
    Button(text="Register", height="2", width="30", command=register).pack()

def register():
    # set global variables
 
    global username
    global password
    global username_entry
    global password_entry
    
    register_screen = Toplevel(main_screen)
    register_screen.title("INCIDENT MANAGEMENT SYSTEM")
    register_screen.geometry("300x250")
    Label(register_screen, text="REGISTER", bg="Blue", width="300", height="2", font=("Calibri", 13)).pack(fill="x")
    Label(register_screen, text="").pack()
    
    # Set text variables
    username = StringVar()
    password = StringVar()
    
    # Set label for user's instruction
    Label(register_screen, text="Please enter details below", bg="Orange").pack()
    Label(register_screen, text="").pack()
    
    # Set username entry
    # The Entry widget is a standard Tkinter widget used to enter or display a single line of text.
    
    username_entry = Entry(register_screen, textvariable=username)
    username_entry.pack()
   
    # Set password label
    password_lable = Label(register_screen, text="Password * ")
    password_lable.pack()
    
    # Set password entry
    password_entry = Entry(register_screen, textvariable=password, show='*')
    password_entry.pack()
    
    Label(register_screen, text="").pack()
    
    # Set register button
    Button(register_screen, text="Register", width=10, height=1, bg="blue").pack()
    
mainGUI()
main_screen.mainloop()
```
