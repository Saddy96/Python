```python
#Entry for the loan amount to be taken
loanAmount = int(input('Enter the Loan Amount to be taken: '))

#Various options for the Rate of interest so as user can select any one of them
roiPeriodType ="""\nWhich interest type u want to select  :
          1. Yearly.
          2. Semiannually
          3. Quartely
          4. Montly
"""

#Variables initialization
begBal = loanAmount
interestRate = 0
totalInterestPaid = 0

#Check if the entered amount is correct or not i.e. Non-zero and Non-negative No
if loanAmount > 0:
    print(roiPeriodType)
    
    #input Entry of the type of the interest type like yearly, semi-annually, etc
    roiPeriodTypeVal = int(input('Please enter your selection from the above option: '))
    
    #Check if the user enters correct selection of Interest type from the given selection
    if roiPeriodTypeVal == 1 or roiPeriodTypeVal == 2 or roiPeriodTypeVal == 3 or roiPeriodTypeVal == 4:
        
        #Entry for Nominal Interest Rate
        nominalInterestRate = int(input('At what interest rate: '))
        
        #Check if the entered nominal interest rate is correct or not i.e. Non-zero and Non-negative No
        if nominalInterestRate > 0:
            
            '''Below conditions will check what user has 
            selected for the interest type then based on that 
            selection the ACTUAL INTEREST RATE will be calculated'''
            if roiPeriodTypeVal == 1:
                finalInterestRate = (nominalInterestRate/100)
            elif roiPeriodTypeVal == 2:
                interestRate = ((((1 + (nominalInterestRate/(2*100)))**2)-1)/100)
                finalInterestRate = round(interestRate,4)
                print(finalInterestRate)
            elif roiPeriodTypeVal == 3:
                interestRate = ((((1 + (nominalInterestRate/(4*100)))**4)-1)/100)
                finalInterestRate = round(interestRate,4)
                print(finalInterestRate)
            elif roiPeriodTypeVal == 4:
                interestRate = ((((1 + (nominalInterestRate/(12*100)))**12)-1)/100)
                finalInterestRate = round(interestRate,4)
                print(finalInterestRate)
             
            #Entry for duration
            duration = int(input('Loan Duration: '))
            
            #Check if the entered duration is correct or not i.e. Non-zero and Non-negative No
            if duration > 0:
                
                #Monthly EMI amount calculation formula
                installments = loanAmount*finalInterestRate * (((1+finalInterestRate)*duration)/(((1+finalInterestRate)*duration) - 1))
                emiInstallments = round(installments,2)
                
                #Doing calculation for all the years
                for i in range(1,duration+1):
                    print('\nBegining Value for',i, 'year is', begBal)
                    print('Your Installments value for every year',i, 'year is', emiInstallments)
                    
                    #Calculating the interest amount to be paid for the respective year
                    interestAmount = round((begBal * finalInterestRate),2)
                    print('Interest amuont for',i, 'year is', interestAmount)
                    
                    #Calculating the Principal paid for the respective year
                    principalAmountPaid = round((emiInstallments - interestAmount),2)
                    print('The principal amount for the EMi for',i, 'year is', principalAmountPaid)
                    
                    #Calculating the remaining Principal amount to be paid after the respective year
                    remmaingBal = begBal - principalAmountPaid
                    begBal = round(remmaingBal,2)
                    print('The Balance Remaining for next is',begBal)
                    
                    #Calculating the total interst amount paid till the end of the respective year
                    totalInterestPaid = round((totalInterestPaid + interestAmount),2)
                
                print('\nTotal Amount paid')
                percentageRatio = round(((totalInterestPaid / loanAmount)*100),2)
                print('You will be paying almost',percentageRatio,'% apart from your loan amount')
                print('\nTotal Interest Amount that u will pay in',duration, 'years will be', totalInterestPaid)
                
            else:
                print('\nCan not calculate for the given periods as the number entered is invalid!!')
                
        else:
            print('\nCan not calculate for the given interest rate as it is less than 0')
            
    else:
        print('\nPlease select a valid option')
        
else:
    print('\n Loan amount can not be less than or equal to 0')
```
