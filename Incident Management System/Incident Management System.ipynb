```python
# importing the modules
import sys

# json Module to handle the data from the text file
import json
```

```python
# Files Operations

# Reading the data from the text file by providing the name of the file
def readFiles(file):
    fileOperation=open(file+".txt")
    textLoginData = fileOperation.read()
    credentailsData = json.loads(textLoginData)
    fileOperation.close()
    return credentailsData

# writing the data to the text file by providing the name of the file and the data
def writeFiles(data,file):
    fileOperation=open(file+".txt",'wt')
    data = json.dumps(data, indent = 4)
    fileOperation.write(data)
    fileOperation.close()
```

```python
# Login Operations

# Verifying the users and the password based on the user data avilable in text file
def registereduser(userId, password):
    registeredUserData = readFiles("Credentials")
    if userId in registeredUserData.keys():
        if password == registeredUserData[userId]:
            print("Login Suscessful!!\n")
            return userId
        else:
            return "unsuccess"

# Getting the different users available in the LoginDetails text file as in actual scenario there will be separate file for the Login
# It will print all the user types available in the LoginDetails file
def usersType():
    print("\nPlease select the type of users you belong to:")
    userTypes = readFiles("LoginDetails")
    countVal = 1
    for key in userTypes.keys():
        print(countVal,".", key)
        countVal += 1
```
