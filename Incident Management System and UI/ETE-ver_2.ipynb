```python
class employee:
    def __init__(self, userName): 
        self.userName = userName
    
    def showUserName(self):
        print(self.userName)

# child class , derived class        
class hubEmployee(employee):
    
    def __init__(self, userName):
        
        # invoking the __init__ of the parent class
        employee.__init__(self, userName)
        self.user = userName
        self.ticketNo = []
        self.raisedBy = []
        self.asignedTo = userName
        self.ticketIssue = []
        self.ticketStatus = []
    
    def reset(self):
        self.ticketNo = []
        self.raisedBy = []
        self.ticketIssue = []
        self.ticketStatus = []
        
    def process(self):
        print("\nWhat do you want to do")
        count = 1
        listData = ["GET TICKETS","ADD TICKETS", "QUIT THE PROCESS"]
        for i in listData:
            print("Choose ",count, "for", i)
            count += 1
    
    # Based on the process function the below function will ask the user to enter their chosen option
    def processSelection(self):
        while(True):
            try:
                selVal = int(input("Enter the selection from above options"))
                if selVal == 1 or selVal == 2 or selVal == 3 or selVal == 4:
                    print("You have selected ",selVal)
                    return selVal
                    break;
                else:
                    print("Please selct only from the above option")
            except ValueError:
                print("\n Not a valid entry, try again.")
            
    def getTickets(self):
        user.reset()
        conn = sqlite3.connect('incidentMS.db')
        query="SELECT * FROM ticktes WHERE asignedTo = '{username}';".format(username = userName)
        c2 = conn.execute(query)
        data = c2.fetchall()
        print("\n Below are the details which are there for you\n")
        print("TICKET NO | RAISED BY |     TICKET ISSUE        | TICKET STATUS")
        for i in data:
            user.ticketNo.append(i[0])
            user.raisedBy.append(i[1])
            user.ticketIssue.append(i[3])
            user.ticketStatus.append(i[4])
            print(i[0],'      |      ',i[1],' |   ',i[3],'       |        ',i[4])
            print("\n")
        
        print("\n")
        
        conn.commit()
        conn.close()

    def insertTickets(self,raisedBy,asignedTo,ticketIssue):
        conn = sqlite3.connect('incidentMS.db')
        
        countQuery="SELECT count(*) FROM ticktes;"
        c1 = conn.execute(countQuery)
        ticketNo = 0
        for col in c1:
            ticketNo = col[0]+1
        
        query="INSERT INTO ticktes (ticketNo,raisedBy,asignedTo,ticketIssue,ticketStatus) \
          VALUES ({},'{}', '{}', '{}','Pending');".format(ticketNo, raisedBy,asignedTo,ticketIssue)

        c2 = conn.execute(query)
        print('Inserted Successfully')
    
        conn.commit()
        conn.close()
        return "INSERTED"
    
    def updateTicket(self, ticketNo, username):
        conn = sqlite3.connect('incidentMS.db')
        query="UPDATE ticktes SET ticketStatus = 'Completed' WHERE ticketNo = {} AND asignedTo ='{}';".format(ticketNo,username)
        c2 = conn.execute(query)

        conn.commit()
        conn.close()
       
```
